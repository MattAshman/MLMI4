```python
import os
import sys
import re
import pdb

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datasets = ['boston', 'concrete', 'energy', 'kin8nm', 'naval', 'power', 'protein', 'wine_red']
```

```python
results_dir = '../regression_results/dsvi_results'
layers = ['2', '3', '4', '5']
dsvi_mean_nlls = np.ndarray(shape=(len(datasets), len(layers)))
dsvi_std_nlls = np.ndarray(shape=(len(datasets), len(layers)))
dsvi_mean_rmse = np.ndarray(shape=(len(datasets), len(layers)))
dsvi_std_rmse = np.ndarray(shape=(len(datasets), len(layers)))
for layer in layers:
    for dataset in datasets:
        filename =  'dsvi_' + dataset + '_' + layer + '_' + '100'
        path = os.path.join(results_dir, filename)
        nll_path = path + '.nll'
        rmse_path = path + '.rmse'
        time_path = path + '.time'
        
        if os.path.isfile(nll_path):
            with open(nll_path, 'r') as f:
                s = f.read()
                results = re.findall('Split\s[0-9]*:\s(.*)\n', s, re.MULTILINE)
            results = [float(result) for result in results if result != 'nan']
            dsvi_mean_nlls[datasets.index(dataset), layers.index(layer)] = np.mean(results)
            dsvi_std_nlls[datasets.index(dataset), layers.index(layer)] = np.std(results)
        else:
            dsvi_mean_nlls[datasets.index(dataset), layers.index(layer)] = np.nan
            dsvi_std_nlls[datasets.index(dataset), layers.index(layer)] = np.nan
            
        if os.path.isfile(rmse_path):
            with open(rmse_path, 'r') as f:
                s = f.read()
                results = re.findall('Split\s[0-9]*:\s(.*)\n', s, re.MULTILINE)
            results = [float(result) for result in results if result != 'nan']
            dsvi_mean_rmse[datasets.index(dataset), layers.index(layer)] = np.mean(results)
            dsvi_std_rmse[datasets.index(dataset), layers.index(layer)] = np.std(results)
        else:
            dsvi_mean_rmse[datasets.index(dataset), layers.index(layer)] = np.nan
            dsvi_std_rmse[datasets.index(dataset), layers.index(layer)] = np.nan
```
